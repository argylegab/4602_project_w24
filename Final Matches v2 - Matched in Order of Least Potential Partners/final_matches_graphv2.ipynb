```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

print("Compared each person in the 1096 x 2 = 2192 against each other 2191 people to see the possible score; found:\n")

with open('final_matches_org.json') as json_file:
    final_matches_data1 = json.load(json_file)

print(str(len(final_matches_data1)) + " People who had the best score with their original partner, so were matched back with them.\n")
print(final_matches_data1)
```

```python
with open('final_matches_one.json') as json_file:
    final_matches_data2 = json.load(json_file)

print(str(len(final_matches_data2)) + " People who only had 1 other person / 2192 that they had a better score with if they were to be a couple\n")
print(final_matches_data2)
```

```python
with open('final_matches_others.json') as json_file:
    final_matches_data3 = json.load(json_file)

print(str(len(final_matches_data3)) + " People who could be matched with someone else.\n")
print(final_matches_data3)
```

```python
# run through each dict record and also put each id in a set but also all rel in dew dict

all_final_matches = {}
all_ids = set()

for record in final_matches_data1:
    all_final_matches[record] = {
        "p_id": final_matches_data1[record]["p_id"],
        "total_score": final_matches_data1[record]['total_score']
    }

    p1_id = int(record)
    p2_id = int((final_matches_data1[record])['p_id'])
    all_ids.add(p1_id)
    all_ids.add(p2_id)
```

```python
for record in final_matches_data2:
    all_final_matches[record] = {
        "p_id": final_matches_data2[record]["p_id"],
        "total_score": final_matches_data2[record]['new_score']
    }

    p1_id = int(record)
    p2_id = int((final_matches_data2[record])['p_id'])
    all_ids.add(p1_id)
    all_ids.add(p2_id)
    
print(len(all_final_matches))
print(len(all_ids))
```

```python
with open('partner_info_dict_json.json') as json_file:
    indiv_attr_data = json.load(json_file)

indiv_data = {}

for record in indiv_attr_data:
    person1_id = int((indiv_attr_data[record])['caseid_new'])
    person2_id = int(record)

    indiv_data[person1_id] = {
        "age": float((indiv_attr_data[record])['w1_ppage']),
        "eth": float((indiv_attr_data[record])['w1_subject_race']),
        "edu": float((indiv_attr_data[record])['w1_ppeduc']),
        "pol": float((indiv_attr_data[record])['w1_partyid7'])
    }

    indiv_data[person2_id] = {
        "age": float((indiv_attr_data[record])['w1_q9']),
        "eth": float((indiv_attr_data[record])['w1_q6b']),
        "edu": float((indiv_attr_data[record])['w1_q10']),
        "pol": float((indiv_attr_data[record])['w1_q12'])
    }

print(len(indiv_data))
```

```python
indiv_id_arr_final_matches = []
partnership_edges_final_matches = []

id_labels_final_matches = []
hover_labels_final_matches = []
titles_final_matches = []
colours_final_matches = []

gradient_colours = ["#cc6600", "#ff004e", "#ff006d", "#ee13a7", "#da35c0", "#9f5ce4", "#4474f0"]

for record in indiv_data:
    person_id = record

    indiv_id_arr_final_matches.append(person_id)
    id_labels_final_matches.append(str(person_id))

    poli = (indiv_data[record])["pol"]
    eth = (indiv_data[record])["eth"]
    age = (indiv_data[record])["age"]
    edu = (indiv_data[record])["edu"]

    person_title = " ".join(["Political Party:", str(poli), "\n",
                          "Race / Eth:", str(eth), "\n",
                          "Age:", str(age), "\n",
                          "Education", str(edu), "\n"])
    
    titles_final_matches.append(person_title)
    colours_final_matches.append(gradient_colours[0])


all_ids_set = set()

for record in indiv_data:
    all_ids_set.add(record)

print(len(all_ids_set))

partnered_ids_set = set()
for record in all_final_matches:
    partnered_ids_set.add(int(all_final_matches[record]['p_id']))
    partnered_ids_set.add(int(record))

print(len(partnered_ids_set))

unpartnered_ids_set = all_ids_set - partnered_ids_set
unpartnered_ids_arr = list(unpartnered_ids_set)

unp_dict = {}

for unp_id in unpartnered_ids_arr:
    unp_dict[unp_id] = indiv_data[unp_id]


all_final_matches[2172] = {
    'p_id': 2047,
    'total_score': 10
}

all_final_matches[2126] = {
    'p_id': 2101,
    'total_score': 10
}


all_final_matches[2038] = {
    'p_id': 1991,
    'total_score': 10
}

all_final_matches[2176] = {
    'p_id': 2180,
    'total_score': 10
}
all_final_matches[1764] = {
    'p_id': 2054,
    'total_score': float(1/13 + 2)
}
all_final_matches[2053] = {
    'p_id': 2071,
    'total_score': float(4+3+float(2/13))
}
all_final_matches[2088] = {
    'p_id': 2142,
    'total_score': float(4+3+2+float(2/13))
}
all_final_matches[2061] = {
    'p_id': 2159,
    'total_score': 10
}

all_final_matches[2042] = {
    'p_id': 2070,
    'total_score': float(4+3+float(2/13))
}

```

```python
for record in all_final_matches:
    partnership_edges_final_matches.append([int(record), int(all_final_matches[record]['p_id']), float(all_final_matches[record]['total_score']), "Score: " + str(all_final_matches[record]['total_score'])])
```

```python
from pyvis.network import Network

net_final = Network()

# add nodes
net_final.add_nodes(
    indiv_id_arr_final_matches,
    label=id_labels_final_matches,
    title=titles_final_matches,
    color=colours_final_matches
)
for edge in partnership_edges_final_matches:
    net_final.add_edge(edge[0], edge[1], value=edge[2], title=edge[3])


net_final.show('finalMatchesv2_pyvisGraph.html', notebook=False)  # save visualization in 'graph.html'
```

```python
# now check score of new matches
all_final_matches_total_scores = {
    "total_pol": 0,
    "total_race": 0,
    "total_age": 0,
    "total_edu": 0,
    "total_all": 0
}

for record in all_final_matches:
    all_final_matches_total_scores["total_all"] += all_final_matches[record]['total_score']


print(all_final_matches_total_scores['total_all'])
print(all_final_matches_total_scores['total_all']/1087)
```
