```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Open the JSON file
with open('total_score_rp_json.json') as json_file:
    # Load JSON data from the file
    total_score_data = json.load(json_file)

# Now 'data' contains the parsed JSON object
print(total_score_data)

with open('score_resp_partner_json.json') as json_file:
    # Load JSON data from the file
    indiv_attr_score_data = json.load(json_file)
print(indiv_attr_score_data)

with open('partner_info_dict_json.json') as json_file:
    # Load JSON data from the file
    indiv_attr_data = json.load(json_file)

# Now 'data' contains the parsed JSON object
print(indiv_attr_data)
```

```python
##############
```

```python
indiv_id_arr = []
partnership_edges = []

id_labels = []
hover_labels = []
titles = []
colours = []


gradient_colours = ["#ff2c2c", "#ff004e", "#ff006d", "#ee13a7", "#da35c0", "#9f5ce4", "#4474f0"]
gradient_poli_steps = ["Strong Republican", "Not Strong Republican", "Leans Republican", "Undecided / Independent", "Leans Democrat", "Not Strong Democrat", "Strong Democrat"]

for record in indiv_attr_data:
    resp_poli = int((indiv_attr_data[record])["w1_partyid7"])
    partner_poli = int((indiv_attr_data[record])["w1_q12"])
    
    if resp_poli != -1:
        resp_id = int(record)
        partner_id = int((indiv_attr_data[record])["caseid_new"])
        indiv_id_arr.append(resp_id)
        indiv_id_arr.append(partner_id)
    
        id_labels.append(str(resp_id))
        id_labels.append(str(partner_id))
    
        partnership_edges.append((resp_id, partner_id))

        colours.append(gradient_colours[resp_poli-1])
        colours.append(gradient_colours[partner_poli-1])

        titles.append(gradient_poli_steps[resp_poli-1])
        titles.append(gradient_poli_steps[partner_poli-1])


        


    
```

```python
from pyvis.network import Network

net = Network()  # create graph object

# add nodes
net.add_nodes(
    indiv_id_arr,  # node ids
    label=id_labels,  # node labels
    # node titles (display on mouse hover)
    title=titles,
    color=colours  # node colors (HEX)
)
# add list of edges, same as in the previous example
net.add_edges(partnership_edges)

net.show('graph.html', notebook=False)  # save visualization in 'graph.html'
```
