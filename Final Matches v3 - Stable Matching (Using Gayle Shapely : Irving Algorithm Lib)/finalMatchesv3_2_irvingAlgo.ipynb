```python
# read in json dictionary of sorted score dictionary
# turn into preference list of just person ids that works with matching library
import json
with open('pref_scores.json') as json_file:
    pref_score = json.load(json_file)
```

```python
pref_dict = {}

for record in pref_score:
  pref_arr = list((pref_score[record]).keys())
  pref_arr_int = []

  for item in pref_arr:
    pref_arr_int.append(item)

  pref_dict[record] = pref_arr_int
```

```python
print("Format of preference list as of now: ", pref_dict["71609"])
```
